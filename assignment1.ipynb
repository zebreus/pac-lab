````typescript
// Load all traces from a file
const loadTraces = async (path: string): Promise<number[][]> => {
    const buffer = (await Deno.readFile(path)).buffer
    const result: Array<Array<number>> = [];
    let offset = 0;
    while (offset < buffer.byteLength) {
      const lengthSlice = new DataView(buffer.slice(offset,offset+4));
      const length = lengthSlice.getUint32(0, true);
      const contentSlice = new Int8Array(buffer.slice(offset+4,offset+4+length))
      result.push(Array.from(contentSlice))
      offset = offset + 4 + length
      // console.log(`Found trace at ${offset} with length ${length}`)
    }
    return result
}

const tracesA = await loadTraces("assignment1/Traces_A.dat")
const tracesB = await loadTraces("assignment1/Traces_B.dat")
````

````typescript
import { plotFunctions } from "./helper.ts";

await plotFunctions({
    name: "Plot for trace A",
    to: tracesA[0].length,
    step: 10,
    xName: "Time",
    yName: "Power Consumption",
    width: 1600,
    data: tracesA.map((trace, index) => [`Trace ${index}`, (n) => trace[n]])
  });
````

In trace A there is one spike in power consumption, then a short pause, and then 11 equally spaced spike. The pause in the beginning is roughly twice as long as the pause between the following spikes. The first spike has a power consumption of around 100, the following spikes only have 95.
